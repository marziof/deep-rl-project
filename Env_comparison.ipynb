# I. Tuning update frequency

```python
DATA_DIR = os.path.join("final_results", "data")

ENV_NAME = "LunarLander"

FILE_PATHS = [
    os.path.join(DATA_DIR, f"{ENV_NAME}_DDQN", f"ddqn_{ENV_NAME}-v3_logs_nep1000.csv"),
    os.path.join(DATA_DIR, f"{ENV_NAME}_DQN", f"dqn_{ENV_NAME}-v3_logs_nep1000.csv"),
    os.path.join(DATA_DIR, f"{ENV_NAME}_PPO_1024", f"ppo_{ENV_NAME}-v3_logs.csv"),
]

dfs = []
for file_path in FILE_PATHS:
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        dfs.append(df)
    else:
        print(f"File not found: {file_path}")

df_combined = pd.concat(dfs, ignore_index=True)
```

```python

SAVE_DIR = os.path.join("final_results", "plots", "LunarLander_comp")
if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)
SAVE_PATH = os.path.join(SAVE_DIR, "LunarLander_comparison.png")
title = "LunarLander Performance"
env_name = "LunarLander-v3"
plot_env_curves(df_combined, metric="eval_reward", env_name=env_name, save_path=SAVE_PATH, bin_size=5)
```
